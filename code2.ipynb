```python
# Assuming df is your DataFrame
# Initialize a dictionary to store the counts of null values for each column containing the pattern 'highbeta'
null_counts = {}

# Iterate over the column names
for col in df.columns:
    # Check if the column name contains the pattern 'highbeta'
    if 'highbeta' in col:
        # Calculate the number of null values in the column
        null_count = df[col].isnull().sum()
        # Store the null count in the dictionary
        null_counts[col] = null_count

# Display the null counts for each column containing the pattern 'highbeta'
for col, count in null_counts.items():
    print("Column '{}': {} null values".format(col, count))
```
